## 创建一个长度为10的0数组

```python
import numpy as np
Z = np.zeros(10)
print(Z)
```

## 创建一个长度为10的0数组，第5个值为1

```python
import numpy as np
Z = np.zeros(10)
Z[4] = 1
print(Z)
```

## 创建一个值从10到49的数组

```python
import numpy as np
Z = np.arange(10,50)
print(Z)
```

## 反转数组

```python
import numpy as np
Z = np.arange(50)
Z = Z[::-1]
print(Z)
```

## 创建一个从0~8的3*3矩阵

```python
import numpy as np
Z = np.arange(9).reshape(3,3)
print(Z)
```

## 找到非0元素的索引

```python
import numpy as np
nz = np.nonzero([1,2,0,0,4,0])
print(nz)
```

## 生成一个3*3的对角矩阵

```python
import numpy as np
Z = np.eye(3)
print(Z)
```

## 生成一个3x3x3的随机数组

```python
import numpy as np
Z = np.random.random((3,3,3))
print(Z)
```
